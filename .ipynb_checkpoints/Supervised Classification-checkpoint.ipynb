* Now, try to train a classifier model to predict the uppercases. Use every single model you know for classification.

```python
# Your code here
data_uppercases = data.sort_values('e', ascending=False)
data_uppercases
```

```python
X = np.array(data.drop(['e'],axis=1))
y = np.array(data['e'])
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.20, random_state=5)
```

```python
logistic_model = linear_model.LogisticRegression()
logistic_model.fit(X,y)
logistic_predict = logistic_model.predict(X)
print(logistic_predict)
```

```python
logistic_model.score(X,y)
```

```python
params={'random_state':0,'max_depth':4}
classifier = DecisionTreeClassifier(**params)
classifier.fit(X_train, y_train)
y_test_pred = classifier.predict(X_test)
y_pred_confussion = classifier.predict(X_test)
confusion_matrix(y_test,y_pred_confussion)
y_pred_proba = classifier.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,y_pred_proba)
auc = metrics.roc_auc_score(y_test,y_pred_proba)
auc
```

```python
classifier= RandomForestClassifier(n_estimators=100,max_depth=4, random_state=0)
classifier.fit(X_train, y_train)
y_test_pred = classifier.predict(X_test)
```
